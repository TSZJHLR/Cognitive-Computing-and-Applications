```python
import spacy

nlp = spacy.load("en_core_web_sm")

def spacy_pos_tag(tokens):
  doc = nlp(" ".join(tokens))
  pos_tags_spacy = [(token.text, token.pos_) for token in doc]
  return pos_tags_spacy

df1['pos_tags_spacy'] = df1['tokens'].apply(spacy_pos_tag)
display(df1)
```

```python
def named_entity_recognition(text):
  doc = nlp(text)
  entities_spacy = [(ent.text, ent.label_) for ent in doc.ents]
  return entities_spacy
```

```python
df1['entities_spacy'] = df1['review'].apply(named_entity_recognition)
display(df1)
```

```python
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
from nltk import ne_chunk, pos_tag, word_tokenize
```

```python
def nltk_ner(text):
  chunks = ne_chunk(pos_tag(word_tokenize(text)))
  entities_nltk = []
  for subtree in chunks.subtrees():
    if subtree.label() != 'S':
      entities_nltk.append((subtree.leaves(), subtree.label()))
  return entities_nltk
```

```python
df1['nltk_entities'] = df1['review'].apply(nltk_ner)
display(df1)
```

```python
print(df1['entities_spacy'].head(3))
print(df1['nltk_entities'].head(3))
```

```python
def get_context(text, target_word):
  doc = nlp(text)
  for token in doc:
    if token.text.lower() == target_word:
      context_spacy = text[token.idx-50:token.idx+50]
      return context_spacy
```

```python
reviews = df1['review'].tolist()[:2]
```

```python
for review in reviews:
  context_spacy = get_context(review, "play")
  if context_spacy:
    print(f"Context: {context_spacy}")
```

```python
def get_context_nltk(text, target_word):
  tokens = word_tokenize(text)
  target_index = -1
  for i, token in enumerate(tokens):
    if token.lower() == target_word:
      target_index = i
      break

  if target_index != -1:
    start_index = max(0, target_index - 5)
    end_index = min(len(tokens), target_index + 6)
    context_nltk = tokens[start_index:end_index]
    return " ".join(context_nltk)
```

```python
reviews = df1['review'].tolist()[:2]
for review in reviews:
  context_nltk = get_context_nltk(review, "play")
  if context_nltk:
    print(f"Context: {context_nltk}")
```

```python
grammar = nltk.CFG.fromstring("""
S -> NP VP
VP -> V NP PP
VP -> V NP
PP -> P NP
NP -> Det N
NP -> Det N PP
NP -> Pronoun
Det -> 'the'
N -> 'man' | 'telescope'
V -> 'saw'
P -> 'with'
Pronoun -> 'I'
""")

parser = nltk.ChartParser(grammar)
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
sia = SentimentIntensityAnalyzer()

def analyze_sentiment(text):
  scores = sia.polarity_scores(text)
  compound_score = scores['compound']
  if compound_score >= 0.05:
    return 'Positive'
  elif compound_score <= -0.05:
    return 'Negative'
  else:
    return 'Neutral'
```

```python
df1['sentiment_analysis'] = df1['review'].apply(analyze_sentiment)

print(df1[['review', 'sentiment_analysis']].head())
```

```python
for index, row in df1.head(5).iterrows():
  scores = sia.polarity_scores(row['review'])
  print(f"Review {index+1}: Sentiment Scores = {scores}")
```

```python
from textblob import TextBlob
```

```python
def analyze_sentiment_spacy(text):
  doc = nlp(text)
  blob = TextBlob(doc.text)
  sentiment = blob.sentiment.polarity
  if sentiment > 0:
    return "Positive"
  elif sentiment < 0:
    return "Negative"
  else:
    return "Neutral"
```

```python
df1['sentiment_analysis'] = df1['review'].apply(analyze_sentiment_spacy)

print(df1[['review', 'sentiment_analysis']].head())
```

```python
for index, row in df1.head(5).iterrows():
  doc = nlp(row['review'])
  blob = TextBlob(doc.text)
  sentiment = blob.sentiment.polarity
  print(f"Review {index + 1}: Sentiment Score = {sentiment}")
```
